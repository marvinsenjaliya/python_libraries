```python
"""
Created on 1/24/2020
@author:Marvin Senjaliya
"""
"""
white wine-quality analysis
"""

import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.express as px
f,axs=plt.subplots(2,2,figsize=(10,7))
read_csv=pd.read_csv('winequality-white.csv',sep=";")
x=read_csv['density']
y=read_csv['quality']
plt.xlabel('density')
plt.ylabel('quality')
plt.title('winequality')
plt.subplot(221)
plt.plot(x,y)
q=read_csv['residual sugar']
w=read_csv['quality']
plt.xlabel('quality')
plt.ylabel('residual sugar')
plt.title('white wine quality')
plt.subplot(222)
sns.barplot(q,w)
e=read_csv['total sulfur dioxide']
r=read_csv['quality']
plt.xlabel('total sulfar dioxied')
plt.ylabel('quality')
plt.subplot(223)
sns.lineplot(e,r)
i=read_csv['fixed acidity']
j=read_csv['quality']
plt.xlabel('fixed acidity')
plt.ylabel('quality')
plt.subplot(224)
sns.barplot(i,j)
```
