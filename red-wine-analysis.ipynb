```python
"""
Created on:1/24/2020
@author:Marvin Senjaliya
"""
"""
red wine analysis
"""
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
df=pd.read_csv('winequality-red.csv',sep=";")
df=df[['density','residual sugar','total sulfur dioxide','fixed acidity','quality']]
df=df.melt('quality',var_name='cols',value_name='vals')
sns.scatterplot(x='vals', y="quality",hue='cols', data=df)
plt.legend()
```
